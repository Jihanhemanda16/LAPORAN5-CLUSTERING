# LAPORAN 5 CLUSTERING

# IMPORT LIBRARY

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
```

# MEMBACA DATASET

```python
data = pd.read_csv("sales.csv")
data
```

# CARI DATA DARI DATASET UNTUK ATRIBUT X

```python
data = data[["W1", "W2"]] #balance x1; purchases 2x
data.head(10)
```

# MELIHAT RINGKASAN STATISTIK

```python
data.describe()
```

```python
fig, ax = plt.subplots()
ax.boxplot(data,
          vert = False,
          showmeans = True,
          meanline = True,
          labels = ("W1", "W2"),
          patch_artist = True,
          medianprops = {"linewidth" : 2, "color" : "red"},
          meanprops = {"linewidth" : 2, "color" : "blue"})
plt.show()
```

# MEMBERSIHKAN OUTLIER

```python
#q1, q3 dan IQR
kolom = ["W1", "W2"]

Q1 = data[kolom].quantile(0.25)
Q3 = data[kolom].quantile(0.75)
IQR = Q3-Q1
data = data[~((data[kolom]<(Q1 - 1.5 * IQR)) |
             (data[kolom]>(Q3 + 1.5 * IQR))).any(axis = 1)]

plt.boxplot(data[kolom])
plt.xticks([1,2], kolom)
plt.title("Outlier Setelah Dibuang")
plt.show()
```

```python
data.describe()
```

```python
data.info()
```

```python
x_array = np.array (data) #nilai X
```

# VISUALISASI PERSEBARAN DATA

```python
plt.scatter(data.W1, data.W2)
plt.show()
```

# STANDARISASI

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_array)
plt.scatter(x_scaled[:,0], x_scaled[:,1], s=10)
plt.show()
```

# CEK SSE K-MEANS CLUSTERING

```python
from sklearn.cluster import KMeans
sse = []
index = range(1,10)
for i in index :
    kmeans = KMeans(n_clusters=i, random_state=0, n_init=10)
    kmeans.fit(x_scaled)
    sse_ = kmeans.inertia_
    sse.append(sse_)
    print (i,sse_)
```

```python
plt.plot(index, sse)
plt.show()
```

# PEMODELAN

```python
kmeans = KMeans (n_clusters = 3, random_state = 0, n_init=10)
kmeans.fit(x_scaled)
```

# CLUSTER PUSAT

```python
kmeans.cluster_centers_
```

# VISUALISASI PERSEBARAN DATA

```python
output = plt.scatter(x_scaled[:,0],x_scaled[:,1], s=10, c = kmeans.labels_)

centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c="red", s=10)
plt.title("KMeans Clustering Sales")
plt.xlabel("W1")
plt.xlabel("W2")
plt.colorbar(output)
plt.show()
```

# EVALUASI MODEL

```python
from sklearn.metrics import davies_bouldin_score
labels = kmeans.labels_
davies_bouldin_score(x_scaled, labels)
```

# VISUALISASI AGGLOMERATIVE HIERARCHICAL

```python
plt.scatter(data['W1'], data['W2'], s=50)
plt.show()
```

```python
data = np.asarray(data[['W1', 'W2']])
print(data)
```

```python
plt.title('Dendrogram')
dendrogram = sch.dendrogram(sch.linkage(data, method='ward'))
```

```python
plt.title('Dendogram')
dendogram = sch.dendrogram(sch.linkage(data, method='average'))
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=5)
output_euclidean = ac.fit_predict(data)

plt.scatter(data[:,0], data[:,1], c=output_euclidean, s=50, cmap='rainbow')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=5, affinity='manhattan', linkage='complete')
output_manhattan = ac.fit_predict(data)

plt.scatter(data[:,0], data[:,1], c=output_manhattan, s=20, cmap='rainbow')
plt.show()
```

```python
from sklearn.metrics import davies_bouldin_score
print(davies_bouldin_score(data, output_euclidean))
print(davies_bouldin_score(data, output_manhattan))
```
